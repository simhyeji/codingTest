```python
from collections import deque
def right(move):
    for m in move:
        m.pop()
        m.insert(0,0)
    return move

def left(move):
    li = []
    for m in move:
        que = deque(m)
        que.popleft()
        que.append(0)
        li.append(list(que))
    return li
def down(move):
    m = len(move)
    move.insert(0,[0]*m)
    move.pop()
    return move

def up(move):
    m = len(move)
    que = deque(move)
    que.popleft()
    que.append([0]*m)
    return list(que)

def upDown(move,m,lock):
    down_li = move
    up_li = move
    for _ in range(m-1):
        down_li = down(down_li)
        if down_li == lock:
            return True
        up_li = up(up_li)
        if up_li == lock:
            return True
    return False

def solution(key, lock):
    #자물쇠 홈과 구멍을 뒤집기
    n= len(lock)
    for lo in lock:
        for i in range(n):
            if lo[i] == 1:
                lo[i] = 0
            else:
                lo[i] = 1

    m = len(key)
    d = n-m #두 길이의 차이
    if d > 0:
        for k in key:
            for _ in range(d):
                k.append(0)
        key.append([[0]*n for _ in range(d)])

    key_li = [key]  #4방향으로 회전한 리스트
    for i in range(3): #왼쪽 방향으로 90도씩 회전
        rot = [[] for _ in range(3)]
        for k in key_li[-1]:  #[[0, 0, 0], [1, 0, 0], [0, 1, 1]]
            for j in range(m):
                rot[j].insert(0,k[j])
        key_li.append(rot)
    #print(key_li)

    for rot in key_li:
        # move_li = rot
        # print(move_li)
        if lock == rot:
            return True
        else: #상하좌우로 이동
            #print("1",rot)
            move = list(rot)
            if upDown(move,m,lock) == True:
                return True
            else:
                #move_li = rot
                #print("2",rot)
                right_li = right(rot)
                #print(right_li)
                c = 0
                while c < m: 
                    
                    if right_li == lock:
                        return True
                    elif upDown(right_li,m,lock) == True:
                        return True
                    else:
                        c += 1
                        right_li = right(right_li)
                        #print(c)
                left_li = left(rot)
                c=0
                while c < m: 
                    if left_li == lock:
                        return True
                    elif upDown(left_li,m,lock) == True:
                        return True
                    else:
                        c += 1
                        left_li = left(left_li)
    return False

key = [[0, 0, 0], [1, 0, 0], [0, 1, 1]]
lock = [[1, 1, 1], [1, 1, 0], [1, 0, 1]]
solution(key,lock)
```

```python
def dfs(v, graph, visited):
    visited[v] = True
    return sum([1] + [dfs(u, graph, visited) for u in graph[v] if not visited[u]])

def solution(n, wires):
    graph = [[] for _ in range(n+1)]
    for v, u in wires:
        graph[v].append(u)
        graph[u].append(v)

    answer = 100
    for i in range(n-1):
        visited = [False for _ in range(n+1)]
        v1, v2 = wires[i]
        visited[v2] = True
        tmp = abs(dfs(v1, graph, visited) - dfs(v2, graph, visited))
        answer = min(tmp, answer)

    return answer
```

```python
import java.util.*;

public class Solution {

    public int bfs(List<List<Integer>> graph, int start, boolean[] visited) {
        Queue<Integer> queue = new LinkedList<>();
        queue.offer(start);
        visited[start] = true;
        int cnt = 0;

        while (!queue.isEmpty()) {
            int v = queue.poll();
            cnt++;
            for (int i : graph.get(v)) {
                if (!visited[i]) {
                    queue.offer(i);
                    visited[i] = true;
                }
            }
        }
        return cnt;
    }

    public int solution(int n, int[][] wires) {
        int answer = n - 2;
        for (int i = 0; i < wires.length; i++) {
            List<List<Integer>> graph = new ArrayList<>();
            for (int j = 0; j < n + 1; j++) {
                graph.add(new ArrayList<>());
            }
            boolean[] visited = new boolean[n + 1];
            int start = wires[i][0];
            int[][] tmps = removeElement(wires, i);
            for (int[] wire : tmps) {
                int x = wire[0];
                int y = wire[1];
                graph.get(x).add(y);
                graph.get(y).add(x);
            }
            int cnt_1 = bfs(graph, start, visited);
            int cnt_2 = n - cnt_1;
            answer = Math.min(answer, Math.abs(cnt_1 - cnt_2));
        }
        return answer;
    }

    private int[][] removeElement(int[][] arr, int index) {
        if (arr == null  index < 0  index >= arr.length) {
            return arr;
        }
        int[][] newArr = new int[arr.length - 1][2];
        int k = 0;
        for (int i = 0; i < arr.length; i++) {
            if (i == index) {
                continue;
            }
            newArr[k++] = arr[i];
        }
        return newArr;
    }
}
```
